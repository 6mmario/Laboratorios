##  1) ¿Cuál es el tamaño en bytes de un array de 1000 elementos de tipo booleano?

```python
# Crear un array de 1000 elementos de tipo booleano
array_bool = np.array([True] * 1000, dtype=bool)

# Calcular el atam;o del array generado
tamanio_bytes = array_bool.nbytes

print(f"El tamaño en bytes del array de 1000 elementos de tipo booleano es: {tamanio_bytes} bytes")
```

## 2) Cree un array cuyos elementos sean los enteros pares en [1,100] y en orden decreciente. Muestre los 10 últimos por pantalla.

```python
# Crear un array con todos los enteros en el rango [1, 100]
# validando que el limite superior debe ser + 1 para ser considerado
todos_los_enteros = np.arange(1, 101)

# Filtrar los números pares
# Utilizando divisi[on modular para validar enteros pares
pares = todos_los_enteros[todos_los_enteros % 2 == 0]

# Ordenar en orden decreciente
pares_decrecientes = np.flip(pares)

# Mostrar los 10 últimos elementos
print("Los 10 últimos elementos del array de enteros pares en orden decreciente son:")
print(pares_decrecientes[-10:])
```

## 4)Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python
# Asignamos los array a las varialbes indicadas
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

'''
Primero se tiene que realizar la resta
de los array elemento por elemento
guadando en una nueva variable
'''
diferencia = a - b

'''
Teniendo la diferencia, usando la funcion de mean
podemos obtener la media aritmetica 
'''
media_diferencia = np.mean(diferencia)

print("La media aritmética de la diferencia (a-b) es:", media_diferencia)
```

```python
# Creamos el array con cordenadas x, y en una variable llamada coordenadas
coordenadas = np.array([[1.33, 4.5],
                        [30.0, 10.7],
                        [70.2, 0.5]])

'''
Para llevar un control de lo que se agrega
creamos una nueva variable
'''
new_coordenada = np.array([37.1, -3.6])
nueva_coordenadas = np.vstack([coordenadas, new_coordenada])

# Mostramos resultado
print("Dimensiones del nuevo array:", nueva_coordenadas.shape)
```

## 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

```python
# Filtramos por granada
granada_municipios = df[df['Provincia'] == 'Granada']

# Dado el resultado, procedemos a realizar los calculos necesarios
media = granada_municipios['Poblacion'].mean()
mediana = granada_municipios['Poblacion'].median()
desviacion_estandar = granada_municipios['Poblacion'].std()
maximo = granada_municipios['Poblacion'].max()
minimo = granada_municipios['Poblacion'].min()

print(f"Media: {media}")
print(f"Mediana {mediana}")
print(f"Desviación estándar: {desviacion_estandar}")
print(f"Valor máximo: {maximo}")
print(f"Valor mínimo: {minimo}")
```

## 6)Realice un histograma con la población de los Municipios para cada una de las provincias.
